```python

import os

# start editable vars #
outputfile_clean= "/home/p.mohammadian.student.sharif/datasets/my_clean.txt"
outputfile_noise= "/home/p.mohammadian.student.sharif/datasets/my_noise.txt"
# file to save the results to
folder_clean= "/home/p.mohammadian.student.sharif/datasets/my_farsdat_training_set/clean"
folder_noise= "/home/p.mohammadian.student.sharif/datasets/my_farsdat_training_set/noise"
# the folder to inventory
exclude= ['*.wav']	# exclude files containing these strings
pathsep= "/"			# path seperator ('/' for linux, '\' for Windows)
# end editable vars #

def list_filenames_of_dir(outputfile,folder,exclude=['*.wav'],pathsep="/"):
    with open(outputfile, "w") as txtfile:
        for path,dirs,files in os.walk(folder):
            sep = "\n---------- " + path.split(pathsep)[len(path.split(pathsep))-1] + " ----------"
            print (sep)
            txtfile.write("%s\n" % sep)

            for fn in sorted(files):
                if not any(x in fn for x in exclude):
                    filename = os.path.splitext(fn)[0]

                    print(filename) 
                    txtfile.write("%s\n" % filename)

    txtfile.close()
```

```python
def list_filepaths_of_dir(outputfile,folder,exclude=['*.wav'],pathsep="/"):
    with open(outputfile, "w") as txtfile:
        for path,dirs,files in os.walk(folder):
            #sep = "\n---------- " + path.split(pathsep)[len(path.split(pathsep))-1] + " ----------"
            #print (sep)
            #txtfile.write("%s\n" % sep)

            for fn in files:
                if not any(x in fn for x in exclude):
                    f = os.path.join(path, fn)

                    print(f) 
                    txtfile.write("%s\n" % f)

    txtfile.close()
```
